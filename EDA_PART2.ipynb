```python
fig,ax=plt.subplots(1,2,figsize=(20,5))

# Countplot is same as barplot except the estimator is explicitly counting the number of occurrences

# The count of the status

print('COUNT PLOT')
sns.set_style('ticks')
sns.countplot(x='status',data=df,ax=ax[0])

# The count of nodes 
sns.countplot(x='nodes',data=df,ax=ax[1])
sns.despine()
```

```python
# Count plot shows the count of observations in each categorical bin using bars

fig,axes=plt.subplots(1,2,figsize=(16,9))

# The count of years in which the surgery is performed
ax1=sns.countplot(x='year',data=df,ax=axes[0])

ax1.set_title("Year count")

# The count of no of nodes that were infected
ax2=sns.countplot(x='nodes',data=df,ax=axes[1])

ax2.set_title("Nodes count")
```
