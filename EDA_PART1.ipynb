```python
#Thsis is a joint plot 
print('JOINT PLOT BETWEEN YEAR AND AGE')
print()
print('The joint plot represents 3 plots')
print('Out of the 3 , one displays a bivariate graph which shows')
print('How the dependent variable(here age) varies with the independent variable(here year)')
print()
print('The plot on top of bivariate graph shows distribution of independent variable(age)')
print()
print('The third plot on right shows the distribution of the dependent variable(year)')
print()
print('We can see there is positive relation between the columns year and age because if the value of one variable increases so does  the other')
print()
print('')
sns.jointplot(x='year',y='age',data=data,kind='hex')
```

```python
print('The data is described in discrete bins')
print()
print('The age of most people is between 40 to 60 ')
print()
sns.distplot(data['age'],kde=False,bins=20,color='red')
```

```python
print('The box plot shows how spreadout the data is ')
print()
print('There are 5 important piece of information')
print('Minimun \nFirst quartile \nThe median \nThird quartile \nThe maximum')
print()
print('The spread can be seen from 40 to 60 age which is same as infered from the distplot')
print('The minimum is about 30')
print('The maximum is above 80')
sns.boxplot(x='year',y='age',data=data)
```

```python
fig=data.corr()
sns.heatmap(fig,cmap='coolwarm')
plt.title('data.corr()')
```

```python
print('Correlation range from -1 to +1\n')
print('If the value is closer to zero means there is no linear trend between the two variables\n ')
print('The close to 1 the correlation is the more positively correlated they are , that is as one increases so does the other and the closer to 1 the stronger this relationship is')
```

```python
print('The people with more number of nodes infected, survived less\n')

print('The people with less number of nodes infected, survived more')

n=sns.FacetGrid(data=data,col='status')
n.map(plt.hist,'nodes')
```

```python
print('The following graphs show the count of age,year and nodes')

fig,ax=plt.subplots(1,3,figsize=(20,5))
#the count of age

sns.countplot(x='age',data=data,ax=ax[0])

ax[0].set_title("Age count")

#the count of years in which the surgery is performed

sns.countplot(x='year',data=data,ax=ax[1])

ax[1].set_title("Year count")

#the count of no of nodes that were infected

sns.countplot(x='nodes',data=data,ax=ax[2])

ax[2].set_title("Nodes count")
```

```python
# Voilin plot similar to box and whiskerplot 
print('The white dot represents the median\n')
print('The thick gray bar in the center represents the interquartile range\n')
print('The thin gray line represents the rest of the distribution, except for points that are determined to be “outliers” using a method that is a function of the interquartile range\n')
sns.violinplot(x="status",y="nodes",data=data,palette='rainbow')
```

```python
print('On each side of the gray line is a kernel density estimation to show the distribution shape of the data. Wider sections of the violin plot represent a higher probability that members of the population will take on the given value; the skinnier sections represent a lower probability.')
print()
print('We can see that the width at 0 is very large implies that maximum patients have 0 nodes infected and have survived more than 5 years')
```

```python
#We can analyse each plot and infer some result out of it
print('You get the bivariate to scatter plots of all the combinations from the variable that is given. Each variable is plotted against the rest of the variables')
print()

a=sns.PairGrid(data,hue='status',palette='rainbow')

a.map(plt.scatter)
```

```python
#as you can see all the inspite of all the data given we still are getting NaN because of the way we represented  

#Pivot table helps you to analyse data and we can easyly see comparisons, patterns and the trends in data
print()
print('PIVOT TABLE')
data.pivot_table(values='age',index='year',columns='nodes')
```

```python
g=data.pivot_table(values='age',index='year',columns='status')
sns.heatmap(g,linecolor='white',linewidth=1)
```
